### Завдання 1

```python
n = 12
N = pow(2, n)
M = 21
NG = 83
i = 0
s2_list = []
while i < N:
    s2 = 2*random.random() + NG*math.cos(2*M*math.pi*i/N)*(1+0.1*random.random()) + 17*math.cos(4*M*math.pi*i/N+random.random()) + 3*math.cos(5*M*math.pi*i/N)*(random.random()+NG)
    s2_list.append(s2)
    i = i + 1
ax = plt.subplots(figsize=(20,3))
ax = plt.plot(s2_list)    
    
```

```python
mean = numpy.mean(s2_list)
var = numpy.var(s2_list)
i = 0
new_s2_list = []
s2_sum = 0
while i < N:
    new_s2_list.append((s2_list[i]-mean)/math.sqrt(var))
    i = i + 1
```

```python
j = 0
R_list = []
while j < N/2:
    i = 0
    temp = 0
    while i < N - j - 1:
        temp = temp + new_s2_list[i]*new_s2_list[i+j]
        i = i + 1
    R_list.append(temp/(N-j-2))
    j = j + 1
```
